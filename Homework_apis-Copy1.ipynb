### Analysis
-We can establish that temperature is directly related to the lattitude position, and the closest the cities are the Ecuador, the warmer are the temperatures.

-Interestingly, humidity stays pretty consistent for the ranges analyzed in the exercize despite the position vs the Ecuador, but with a slight increase as it moves farther.

-Cloudiness and wind speed have a very similar behavior, which makes sense since the slower are the winds, the more the chances of stationary clouds.


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import itertools
import numpy as np
import requests
import time
from config import api_key                    #import openweathermapy.core as owm
from citipy import citipy                     # Import API key from config import gkey
                                              # Incorporated citipy to determine city based on latitude and longitude
output_data_file = "output_data/cities.csv"   # Output File (CSV)
lat_range = (-90, 90)                         #Range of latitudes 
lng_range = (-180, 180)                       #Range of longitudes
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units="metric"
query_url = f"{url}appid={api_key}&units={units}&q="

lat = []
temp=[]
lng=[]
humidity=[]
clouds=[]
winds=[]
#city_list=[]
citylist=[]

for count,city in enumerate(cities, start=1):
    response=requests.get(query_url + city).json()
    print("Processing Record ",count,": " + city)
    print (query_url + city)
    try:      
        lat.append(response['coord']['lat'])
        temp.append(response['main']['temp'])
        lng.append(response['coord']['lon'])
        humidity.append(response['main']['humidity'])
        clouds.append(response['clouds']['all'])
        winds.append(response['wind']['speed'])
        citylist.append(city)    
    except:
        print(f"Values not found for City" + city)
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#Creating dataframe
weather_data = pd.DataFrame({"City": citylist,"Lattitude": lat,"Longitud":lng,"Temperature": temp,
                             "Humidity":humidity,"Cloudiness":clouds,"Wind Speed":winds})
#Export and display dataframe
weather_data.to_csv(r'output_data_file')
weather_data.head()
```
